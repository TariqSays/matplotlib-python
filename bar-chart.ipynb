```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from matplotlib import style
```

```python
courses = ['AIC','CNC','IOT','BCC','QCC']
c1 = [22,34,11,43,32]
c2 = [42,24,51,33,42]
c3 = [32,34,43,23,49]
```

```python
plt.bar(courses,c1)
```

```python
plt.barh(courses,c1)
```

```python
plt.bar(courses,c1,width=0.6)
```

```python
plt.bar(courses,c1,width=0.6, color='red')
```

```python
plt.bar(courses,c1,width=0.6, color='red', edgecolor='blue')
```

```python
plt.bar(courses,c1,width=0.6, color='red', edgecolor='blue', linewidth=5)
```

```python
plt.bar(courses,c1,width=0.6, color='red', edgecolor='blue', linewidth=5, alpha=0.7)
```

```python
plt.bar(courses,c1,width=0.6, color='red', edgecolor='blue', linewidth=5, alpha=0.7, linestyle='--')
```

```python
plt.bar(courses,c1,width=0.6, color='red', edgecolor='blue', linewidth=5, alpha=0.7,label='Class Attendance')
plt.legend()
plt.show()
```

```python
style.use('ggplot')
plt.bar(courses,c1,width=0.6, color='red', edgecolor='blue', linewidth=5, alpha=0.7,label='Class Attendance')
plt.legend()
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize=(13,5))
plt.bar(courses,c1,width=0.6, color='red', edgecolor='blue', linewidth=5, alpha=0.7,label='Class Attendance')
plt.legend()
plt.show()
```

```python
style.use('ggplot')
plt.barh(courses,c1, color='red', edgecolor='blue', linewidth=5, alpha=0.7,label='Class Attendance')
plt.legend()
plt.show()
```

```python
style.use('ggplot')
plt.bar(courses,c1,width=0.6, color='red', edgecolor='blue', linewidth=5, alpha=0.7,label='Class Attendance')
plt.title('Class Attendance Bar Chart')
plt.xlabel('Number of Students')
plt.ylabel('Courses')
plt.legend()
plt.show()
```

```python
style.use('ggplot')
plt.bar(courses,c1,width=0.6, color='red', edgecolor='blue', linewidth=5, alpha=0.7,label='Class Attendance')
plt.title('Class Attendance Bar Chart',fontsize=15)
plt.xlabel('Number of Students',fontsize=15)
plt.ylabel('Courses',fontsize=15)
plt.legend()
plt.show()
```

```python
style.use('ggplot')
plt.bar(courses,c1,width=0.4, color='red', edgecolor='blue', label='Class Attendance')
plt.bar(courses,c2,width=0.4, color='yellow', edgecolor='green', label='Class Attendance')
plt.title('Class Attendance Bar Chart',fontsize=15)
plt.xlabel('Number of Students',fontsize=15)
plt.ylabel('Courses',fontsize=15)
plt.legend()
plt.show()
```

```python
style.use('ggplot')
plt.bar(courses,[c1,c2], color=['red','green'], edgecolor=['blue','yellow'], label='Class Attendance')
plt.title('Class Attendance Bar Chart',fontsize=15)
plt.xlabel('Number of Students',fontsize=15)
plt.ylabel('Courses',fontsize=15)
plt.legend()
plt.show()
```

```python
style.use('ggplot')
class_index = np.arange(len(courses))
width = 0.4
plt.figure(figsize=(9,5))
plt.bar(class_index,c1,width, color='red', edgecolor='blue', label='Class Attendance')
plt.bar(class_index+width,c2,width, color='yellow', edgecolor='green', label='Class Attendance')
plt.xticks(class_index+width,courses, rotation=35)
plt.title('Class Attendance Bar Chart',fontsize=15)
plt.xlabel('Number of Students',fontsize=15)
plt.ylabel('Courses',fontsize=15)
plt.legend()
plt.show()
```

```python
style.use('ggplot')
class_index = np.arange(len(courses))
width = 0.3
plt.figure(figsize=(9,5))
plt.bar(class_index,c1,width, color='red', edgecolor='blue', label='Class Attendance')
plt.bar(class_index+width,c2,width, color='yellow', edgecolor='green', label='Class Attendance')
plt.bar(class_index+width+width,c3,width, color='green', edgecolor='green', label='Class Attendance')
plt.xticks(class_index+width,courses, rotation=35)
plt.title('Class Attendance Bar Chart',fontsize=15)
plt.xlabel('Number of Students',fontsize=15)
plt.ylabel('Courses',fontsize=15)
plt.legend()
plt.show()
```

```python
style.use('ggplot')
class_index = np.arange(len(courses))
width = 0.3
plt.figure(figsize=(9,5))
plt.barh(class_index,c1,width, color='red', edgecolor='blue', label='Class Attendance')
plt.barh(class_index+width,c2,width, color='yellow', edgecolor='green', label='Class Attendance')
plt.barh(class_index+width+width,c3,width, color='green', edgecolor='green', label='Class Attendance')
plt.yticks(class_index+width,courses, rotation=35)
plt.title('Class Attendance Bar Chart',fontsize=15)
plt.ylabel('Number of Students',fontsize=15)
plt.xlabel('Courses',fontsize=15)
plt.legend()
plt.show()
```
