```python
x = df['Rating']
y = df['Reviews']
```

```python
x.head()
```

```python
y.head()
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.title('Google PlayStore Dataset')
plt.xlabel('Ratings')
plt.ylabel('Reviews')
plt.show()
```

```python
plt.scatter(x,y,c='r')
plt.title('Google PlayStore Dataset')
plt.xlabel('Ratings')
plt.ylabel('Reviews')
plt.show()
```

```python
plt.figure(figsize=(9,5))
plt.scatter(x,y,c='green')
plt.title('Google PlayStore Dataset')
plt.xlabel('Ratings')
plt.ylabel('Reviews')
plt.show()
```

```python
plt.figure(figsize=(9,5))
plt.scatter(x,y,c='green', marker='*', s=120)
plt.title('Google PlayStore Dataset')
plt.xlabel('Ratings')
plt.ylabel('Reviews')
plt.show()
```

```python
plt.figure(figsize=(9,5))
plt.scatter(x,y,c='green', marker='o', s=120)
plt.title('Google PlayStore Dataset')
plt.xlabel('Ratings')
plt.ylabel('Reviews')
plt.show()
```

```python
plt.figure(figsize=(9,5))
plt.scatter(x,y,c='green', marker='o', s=120, alpha=0.7)
plt.title('Google PlayStore Dataset')
plt.xlabel('Ratings')
plt.ylabel('Reviews')
plt.show()
```

```python
plt.figure(figsize=(9,5))
plt.scatter(x,y,c='green', marker='o', s=120, alpha=0.7,edgecolor='r')
plt.title('Google PlayStore Dataset')
plt.xlabel('Ratings')
plt.ylabel('Reviews')
plt.show()
```
