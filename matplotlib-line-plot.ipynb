```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
days = [1,2,3,4,5,6]
temperature = [34.5,43.5,39.6,44.5,34.5,35.6]
plt.plot(days,temperature)
plt.show()
```

```python
days = [1,2,3,4,5,6]
temperature = [34.5,43.5,39.6,44.5,34.5,35.6]
plt.plot(days,temperature)
plt.title("Lodhran Temperature")
plt.xlabel('Days')
plt.ylabel("Temperature")
plt.show()
```

```python
days = [1,2,3,4,5,6]
temperature = [34.5,43.5,39.6,44.5,34.5,35.6]
plt.plot(days,temperature, color='red')
plt.title("Lodhran Temperature")
plt.xlabel('Days')
plt.ylabel("Temperature")
plt.show()
```

```python
days = [1,2,3,4,5,6]
temperature = [34.5,43.5,39.6,44.5,34.5,35.6]
plt.plot(days,temperature, color='green', marker='o')
plt.title("Lodhran Temperature")
plt.xlabel('Days')
plt.ylabel("Temperature")
plt.show()
```

```python
days = [1,2,3,4,5,6]
temperature = [34.5,43.5,39.6,44.5,34.5,35.6]
plt.plot(days,temperature, color='green', marker='o', linestyle=':')
plt.title("Lodhran Temperature")
plt.xlabel('Days')
plt.ylabel("Temperature")
plt.show()
```

```python
days = [1,2,3,4,5,6]
temperature = [34.5,43.5,39.6,44.5,34.5,35.6]
plt.plot(days,temperature, color='green', marker='o', linestyle=':', linewidth=3)
plt.title("Lodhran Temperature")
plt.xlabel('Days')
plt.ylabel("Temperature")
plt.show()
```

```python
days = [1,2,3,4,5,6]
temperature = [34.5,43.5,39.6,44.5,34.5,35.6]
plt.plot(days,temperature, color='green', marker='o', linestyle=':', linewidth=3, markersize=22)
plt.title("Lodhran Temperature")
plt.xlabel('Days')
plt.ylabel("Temperature")
plt.show()
```
