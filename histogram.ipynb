```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
ml = np.random.randint(18,45,(100))
py = np.random.randint(15,35,(100))
```

```python
print(ml)
print(py)
```

```python
plt.hist(ml)
plt.show()
```

```python
plt.hist(ml)
plt.title('Machine Learning Students Histogram')
plt.show()
```

```python
plt.hist(ml)
plt.title('Machine Learning Students Histogram')
plt.xlabel('Age Classes')
plt.ylabel('Number of Students')
plt.show()
```

```python
bins = [15,18,21,24,27,30,33,36,39,42,45]
plt.hist(ml,bins)
plt.title('Machine Learning Students Histogram')
plt.xlabel('Age Classes')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.hist(ml, rwidth=0.7)
plt.title('Machine Learning Students Histogram')
plt.xlabel('Age Classes')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.hist(ml, histtype='step')
plt.title('Machine Learning Students Histogram')
plt.xlabel('Age Classes')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.hist(ml, histtype='step', align='mid')
plt.title('Machine Learning Students Histogram')
plt.xlabel('Age Classes')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.hist(ml,orientation='horizontal',rwidth=0.8)
plt.title('Machine Learning Students Histogram')
plt.xlabel('Age Classes')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.hist(ml,orientation='horizontal',rwidth=0.8, color='y')
plt.title('Machine Learning Students Histogram')
plt.xlabel('Age Classes')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.hist(ml,orientation='horizontal',rwidth=0.8,color='m',label='Students Age Classification')
plt.title('Machine Learning Students Histogram')
plt.xlabel('Age Classes')
plt.ylabel('Number of Students')
plt.legend()
plt.show()
```

```python
plt.hist(ml,orientation='horizontal',rwidth=0.8,color='m',label='ML Students')
plt.hist(py, orientation='horizontal', rwidth=0.8, color='r',label='Python Students')
plt.title('Machine Learning Students Histogram')
plt.xlabel('Age Classes')
plt.ylabel('Number of Students')
plt.legend()
plt.show()
```

```python
plt.hist(ml,orientation='vertical',rwidth=0.8,color='m',label='ML Students')
plt.hist(py, orientation='horizontal', rwidth=0.8, color='r',label='Python Students')
plt.title('Machine Learning Students Histogram')
plt.xlabel('Age Classes')
plt.ylabel('Number of Students')
plt.legend()
plt.show()
```

```python
plt.hist(ml,orientation='vertical',rwidth=0.8,color='m',label='ML Students')
plt.hist(py, orientation='vertical', rwidth=0.8, color='r',label='Python Students')
plt.title('Machine Learning Students Histogram')
plt.xlabel('Age Classes')
plt.ylabel('Number of Students')
plt.legend()
plt.show()
```

```python
plt.hist([ml,py],orientation='vertical',rwidth=0.8,color=['r','b'],label=['ML Students','Python Students'])
#plt.hist(py, orientation='vertical', rwidth=0.8, color='r',label='Python Students')
plt.title('Machine Learning Students Histogram')
plt.xlabel('Age Classes')
plt.ylabel('Number of Students')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.hist([ml,py],orientation='vertical',rwidth=0.8,color=['r','b'],label=['ML Students','Python Students'])
#plt.hist(py, orientation='vertical', rwidth=0.8, color='r',label='Python Students')

plt.title('Machine Learning Students Histogram')
plt.xlabel('Age Classes')
plt.ylabel('Number of Students')
plt.show()
```
