```python
import matplotlib.pyplot as plt
```

```python
courses = ['CNC','AIC','IOT','BCC','QNC']
c1 = [43,54,34,23,45]
```

```python
plt.pie([1])
plt.show()
```

```python
plt.pie(c1)
plt.show()
```

```python
plt.pie(c1, labels=courses)
plt.show()
```

```python
explode = [0,0,0,0,0]
plt.pie(c1, labels=courses, explode = explode)
plt.show()
```

```python
explode = [0,0,0.1,0,0]
plt.pie(c1, labels=courses, explode = explode)
plt.show()
```

```python
explode = [0,0,0.1,0,0]
color = ['c','b','r','y','g']
plt.pie(c1, labels=courses, explode = explode, colors = color)
plt.show()
```

```python
explode = [0,0,0.1,0,0]
color = ['c','b','r','y','g']
plt.pie(c1, labels=courses, explode = explode, colors = color, autopct='%0.1f%%')
plt.show()
```

```python
explode = [0,0,0.1,0,0]
color = ['c','b','r','y','g']
plt.pie(c1, labels=courses, explode = explode, colors = color, autopct='%0.1f%%', shadow=True)
plt.show()
```

```python
explode = [0,0,0.1,0,0]
color = ['c','b','r','y','g']
plt.pie(c1, labels=courses, explode = explode, colors = color, autopct='%0.1f%%', shadow=True, radius=1.7)
plt.show()
```

```python
explode = [0,0,0.1,0,0]
color = ['c','b','r','y','g']
plt.pie(c1, labels=courses, explode = explode, colors = color, autopct='%0.1f%%', shadow=True, startangle=270,radius=1.7)
plt.show()
```

```python
explode = [0,0,0.1,0,0]
color = ['c','b','r','y','g']
textprops = {'fontsize':15}
plt.pie(c1, labels=courses, explode = explode, colors = color, autopct='%0.1f%%', shadow=True, startangle=270,radius=1.7, textprops = textprops)
plt.show()
```
